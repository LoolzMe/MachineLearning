```python
import tensorflow as tf
import numpy as np
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
import tensorflow_probability as tfp

print(tf.__version__)
```

```python
d0 = Dense(units=1, input_shape=[1])
model = Sequential([d0])
model.compile(optimizer="sgd", loss='mean_squared_error')

X = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0, 6.0, 8.0], dtype=float)
Y = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0, 11.0, 15.0], dtype=float)

# real function: Y = 2X - 1
print(tfp.stats.correlation(X,Y, sample_axis=0, event_axis=None))
```

```python
model.fit(X, Y, epochs=700)
```

```python
k = 893
predicted = int(tf.math.round(model.predict([k])))

print("real: ", 2*k - 1)
print("predicted: ", predicted)
```
